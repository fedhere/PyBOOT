```python
##second degree polynomial: arguments are i
#independent variable, dependent variable, degree of the polynomial

x = np.arange(0, 3, .02)
y1, y2 = 2*x**2, 2*x**2.05


c = np.exp(x)
d = c[::-1]

pl.plot(x, y1, 'o', color='DarkGreen', label='one line ')
pl.plot(x, y2, 'o', color='DarkGreen', label='two lines ')
pl.plot(x, c, 'k--', label='Model')
pl.plot(x, d, 'k-', label='wrong model')
pl.plot(x, c + d, 'k-', label='models sum')



#pl.plot(xnew, np.poly1d(mrnew)(xnew), 'y', alpha=1)
legend = pl.legend(loc='upper right', shadow=True, fontsize='x-large')


pl.xlabel("some IV (in fed's units)")
pl.ylabel("some DVs (in fed's units)")
```

```python
#a histogram. there are many choices to make!
#(also see http://matplotlib.org/1.3.0/examples/pylab_examples/histogram_demo_extended.html)
```

```python
fig = pl.figure()
ax = fig.add_subplot(111)
h, b, p = pl.hist([myline, myline*2])

ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
```

```python
fig = pl.figure()
ax = fig.add_subplot(111)
h, b, p = pl.hist([myline, myline*2], histtype='stepfilled', alpha=0.3)
pl.title("overlapping bins, with opacity reduced for visibility")

fig = pl.figure()
ax = fig.add_subplot(111)
h, b, p = pl.hist([myline, myline*2], histtype='step')
pl.title("step only bins (typically confusing)")

fig = pl.figure()
ax = fig.add_subplot(111)
h, b, p = pl.hist([myline, myline*2], histtype='step', stacked=True, fill=True, alpha=0.3)
pl.title("whatch out for stacked bins: they can go very wrong!")


fig = pl.figure()
ax = fig.add_subplot(111)
h, b, p = pl.hist([myline, myline*2], histtype='bar', stacked=True, fill=True)
pl.title("this is a stacked histogram, good for complementary percantages")

fig = pl.figure()
ax = fig.add_subplot(111)
h, b, p = pl.hist([myline, myline*2], histtype='bar', stacked=False, fill=True , alpha=1)
pl.title("these are the same bins but side by side")

#http://matplotlib.org/1.3.0/examples/pylab_examples/histogram_demo_extended.html
```

# Kelly colors for compliance with color blindness

```python
from cycler import cycler

kelly_colors_hex = [
    '#FFB300', # Vivid Yellow
    '#803E75', # Strong Purple
    '#FF6800', # Vivid Orange
    '#A6BDD7', # Very Light Blue
    '#C10020', # Vivid Red
    '#CEA262', # Grayish Yellow
    '#817066', # Medium Gray
    '#007D34', # Vivid Green
    '#F6768E', # Strong Purplish Pink
    '#00538A', # Strong Blue
    '#FF7A5C', # Strong Yellowish Pink
    '#53377A', # Strong Violet
    '#FF8E00', # Vivid Orange Yellow
    '#B32851', # Strong Purplish Red
    '#F4C800', # Vivid Greenish Yellow
    '#7F180D', # Strong Reddish Brown
    '#93AA00', # Vivid Yellowish Green
    '#593315', # Deep Yellowish Brown
    '#F13A13', # Vivid Reddish Orange
    '#232C16', # Dark Olive Green
    ]


newparams = {
  "lines.linewidth": 2.0,
  "axes.edgecolor": "#aaaaaa",
  "patch.linewidth": 1.0,
  "legend.fancybox": 'false',
  'axes.prop_cycle': cycler('color', kelly_colors_hex),
  "axes.facecolor": "#ffffff",
  "axes.labelsize": "large",
  "axes.grid": 'false',
  "patch.edgecolor": "#555555",
  "axes.titlesize": "x-large",
}
pl.rcParams.update(newparams)
```
