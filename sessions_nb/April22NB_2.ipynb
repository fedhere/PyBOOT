```python
keywords = {}

keywords["system"] = ["airmass", "seeingEff", "HA", "moonDistance", 
            "solarElong", "saturation_mag", "AngleMean", "slew", 
            "filter changes", "OpenShutterFraction", "fiveSigmaDepth",
            "CoaddM5", "N Seasons", "NSeasons", "Angle", "rotSkyPos",
            "skyBrightness", "Area", "ExgalM5", "Parallax", "Proper Motion", 
            "fONv", ]
keywords["general"] = ["NumberOfQuickRevisits", "SurfaceBrightLimit",
                        "Teff", "Nvisits"
                        "Unique Nights",
                       "Season Length",  "Fraction of visits"]

keywords["pops"] = ["SNNSNMetric_nSN", "Kuiper", "AGN", 
                     "Brown Dwarf", "SNSLMetric", "young_stars",
                    'KNe', "Microlensing", "SNNSN", "TDE", "XRBPopMetric",
                    "QSO", "FractionPop", "TDC", "PeriodicStarModulationMetric",
                    "WeakLensingNvisits"
                    ]

keywords["Galplane"] = ["Galplane"]

keywords["counts"] = ["N Galaxies", "N Stars", "NgalScale", "Nobj",
                      "NlcPoints", "NVisits", "Unique Nights"]
                      
keywords["lightcurves"] = ["pairs in", 
            "detections in", "NlcPoints", "PeriodDetection",
            "Tgaps", "Inter-Night Gap", "Intranight",  
                       "Intra-Night Gap", "Internight",]

keywords["ddfs"] = ["ELAISS1", "XMM-LSS", "ECDFS", "COSMOS", "EDFS"]



for c in allruns.columns:
  if not np.array([k.lower() in c.lower() for kk in keywords.keys() for k in kk]).any():
    print(c)

```

```python
statdvalues = []
for c in allruns.columns:
  if c.split()[0] in ["Min", "Max", "Median", "Mean", "Rms"]:
    if not ' '.join(c.split()[1:]) in statdvalues:
      print (' '.join(c.split()[1:]))
      statdvalues.append(' '.join(c.split()[1:]))
```
